```python
#Problem 1
import numpy as np
import matplotlib.pyplot as plt
t=np.linspace(0,6*np.pi,300)
x=4*t*np.cos(5*t)
y=t*np.sin(5*t)
z=t


plt.subplot(2,1,1)
plt.plot(x,y,'or-')
plt.tight_layout()
plt.xlabel('x')
plt.ylabel('y')


import matplotlib.pyplot as plt 
from mpl_toolkits.mplot3d import Axes3D
ax=plt.subplot(2,1,2,projection = '3d')
ax.plot(x,y,z,'om-')
plt.tight_layout()
plt.xlabel('x')
plt.ylabel('y')
ax.set_zlabel('z')
```

```python
#Problem 2
import numpy as np
import matplotlib.pyplot as plt
Ta=np.arange(253,325,np.pi/6)
R=8.314 #J/Mol*K Gas Constant
A=7*10**16 #the preexponential(or frequency)factor
E=1*10**5 #activation energy (J/mol)
k=A*np.exp(-E/(R*Ta)) #reaction rate s^-1
#print(k)
plt.subplot(1,2,1)
plt.plot(Ta,k,'g')
plt.tight_layout()
plt.xlabel('x')
plt.ylabel('y')
plt.title('k versus Ta')

plt.subplot(1,2,2)
plt.plot(1/Ta,np.log10(k),'r')
plt.tight_layout()
plt.xlabel('x')
plt.ylabel('y')
plt.title('log10k versus 1/Ta')
```
