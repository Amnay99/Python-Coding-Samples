```python
temp_file=open('Temperature.txt','r')
firstline=temp_file.readline()
import numpy as np

data=np.fromfile(temp_file,dtype='float',count=12*4,sep=" ")
temp_file.close()
#print(data)

data.shape=(12,4)
#print(data)


import matplotlib.pyplot as plt
plt.plot(data[:,0],data[:,1],'rd',label='high')
plt.plot(data[:,0],data[:,2],'ob',label='low')
plt.plot(data[:,0],data[:,3],'xg',label='ave')
plt.legend()
plt.xlabel('Month')
plt.ylabel('Temperature (F)')
```
