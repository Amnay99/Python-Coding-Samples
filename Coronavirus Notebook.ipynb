```python
import pandas as pds

import matplotlib.pyplot as plt

#Data for percentage of countries contribution to CoronaVirus cases

df = [31.050114060147138,0.4873562333480892,68.46252970650477];

 

# Create a pandas series

series = pds.Series(df, index=("USA", "South Korea", "Other Countries"));

 

# Create a pie chart

series.plot.pie(label="", title="Contribution Percentage of WorldWide Cases");

 

plt.show(block=True);

print(USA)
print(SKorea)
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
height = [SKorea_Test, USA_Test]
bars = ('South Korea', 'USA',)
y_pos = np.arange(len(bars))
 
# Create bars
plt.bar(y_pos, height, color=['black','blue'])
plt.title('Testing Per 1 Million of Population of Each Country')
plt.xlabel('Countries')
plt.ylabel('values')
 
# Create names on the x-axis
plt.xticks(y_pos, bars)
 
# Show graphic
plt.show()
```
