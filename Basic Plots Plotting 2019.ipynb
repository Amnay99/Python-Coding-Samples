```python
import numpy as np
import matplotlib.pyplot as plt
#create points for sine function
x = np.linspace(0,2*np.pi,30)
y = np.sin(x)
y1=np.cos(x)
#print(x)
#print(y)

#subplot function indicates how many 
#plots on one figure
plt.subplot(2,1,1)
plt.xlabel('x')
plt.ylabel('y')
plt.axis([0,2*np.pi,-1,1])
plt.plot(x,y,'ro-',label='sine')
plt.legend()

plt.subplot(2,1,2)
plt.xlabel('x')
plt.ylabel('y')
plt.axis([0,2*np.pi,-1,1])
plt.plot(x,y1,'bs-',label='cosine')
plt.legend()



```
